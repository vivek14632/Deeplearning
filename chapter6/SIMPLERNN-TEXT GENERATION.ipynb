```python

#This is a program made using SimpleRNN which can predict the text given 10 previous characters and dataset is based on a book Alice in Wonderland
#from _future_ import print_function----->not needed, was giving some error
#training method used in this code is slightly different than what we were doing right now. In this code, we are training a model for an epoch and then testing it for each epoch testing it

#importing necessary modules
#from_future_ import print_function ----> not needed / confusion
from keras.layers import Dense,Activation
from keras.layers.recurrent import SimpleRNN
from keras.models import Sequential
#from keras.utils.visualize_util import plot  -----> not needed / confusion
import numpy as np
fin=open("Alice.txt",'rb') #open the txt file in binary mode
lines=[]
for line in fin:
    line=line.strip().lower()
    line=line.decode("ascii","ignore")
    if len(line)==0:
        continue
    lines.append(line)
fin.close()
text=" ".join(lines)


```

```python
text
```

```python
#creating lookup tables, this code will create a dictionary
chars=set([c for c in text])
nb_chars=len(chars)
char2index=dict((c,i)for i,c in enumerate(chars))
index2char=dict((i,c)for i,c in enumerate(chars))

index2char
```

```python
char2index
```

```python

#creating input and label texts
SEQLEN = 10
STEP=1
input_chars=[]
label_chars=[]
for i in range(0,len(text)-SEQLEN,STEP):
    input_chars.append(text[i:i+SEQLEN])
    label_chars.append(text[i+SEQLEN])
"""
the following with the preceding code will look like this
he is a handsome man==>
he is a ha-->n
e is a han-->d
 is a hand-->s
is a hands-->o
s a handso-->m
 a handsom-->e
a handsome--> 
 handsome -->m
handsome m-->a
andsome ma-->n

"""
input_chars
```

```python
label_chars
```

```python
SEQLEN
```

```python
nb_chars
```

```python
#one hot encoding done in the following lines of code. Each input is OHE vector of size (nb_chars)
x=np.zeros((len(input_chars),SEQLEN,nb_chars),dtype=np.bool)
y=np.zeros((len(input_chars),nb_chars),dtype=np.bool)
for i,input_chars in enumerate(input_chars):
    for j,ch in enumerate(input_chars):
        x[i,j,char2index[ch]]=1
    y[i,char2index[label_chars[i]]]=1
    
    
    
#Hyper Parameter declaration, look for more details in "deep learning with keras" pg 180
HIDDEN_SIZE=128
BATCH_SIZE=128
NUM_ITERATIONS=25
NUM_EPOCH_PER_ITERATION=1
NUM_PREDS_PER_EPOCH=100
model=Sequential()
model.add(SimpleRNN(HIDDEN_SIZE,input_shape=(SEQLEN,nb_chars),return_sequences=False))#not sure whether to use unroll or not???
model.add(Dense(nb_chars))
model.add(Activation("softmax"))
model.compile(loss="categorical_crossentropy",optimizer="rmsprop")



#training approach
for iteration in range(NUM_ITERATIONS):
    print("="*50)
    print("iteration number : %d"%(iteration))
    model.fit(x,y,batch_size=BATCH_SIZE,epochs=NUM_EPOCH_PER_ITERATION)
    
    test_idx=np.random.randint(len(input_chars))
    test_chars=input_chars[test_idx]
    print("Generating from seed: %s"%(test_chars))
    print(test_chars,end="")
    for i in range(NUM_PREDS_PER_EPOCH):
        Xtest=np.zeros((1,SEQLEN,nb_chars))
        for i,ch in enumerate(test_chars):
            Xtest[0,i,char2index[ch]]=1
        pred=model.predict(Xtest,verbose=0)[0]
        ypred=index2char[np.argmax(pred)]
        #move forward with test_char+ypred
        test_chars=test_chars[1:]+ypred
        print(test_chars)
#some slight error here, please check this and if anybody have some idea to correct, please suggest this.
```
